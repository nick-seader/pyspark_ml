# Decision Trees, Random Forests, and Gradient Boosted Trees

```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('tree').getOrCreate()
```

```
data = spark.read.csv('/FileStore/tables/College.csv', inferSchema=True, header=True)
```

```
data.printSchema()
```

```
data.head(1)[0]
```

# Process Data

```
from pyspark.ml.feature import VectorAssembler
```

```
data.columns
```

```
input_list = [
    'Apps', 
    'Accept', 
    'Enroll', 
    'Top10perc', 
    'Top25perc', 
    'F_Undergrad', 
    'P_Undergrad', 
    'Outstate',
    'Room_Board',
    'Books',
    'Personal',
    'PhD',
    'Terminal',
    'S_F_Ratio',
    'perc_alumni',
    'Expend',
    'Grad_Rate'
]
```

```
assembler = VectorAssembler(
    inputCols=input_list, 
    outputCol='features'
)
```

```
output = assembler.transform(data)
```

```
from pyspark.ml.feature import StringIndexer
```

```
indexer = StringIndexer(inputCol='Private', outputCol='PrivateIndex')
```

```
output_fixed = indexer.fit(output).transform(output)
```

```
output_fixed.printSchema()
```

```
final_data = output_fixed.select('features', 'PrivateIndex')
```

```
train_data, test_data = final_data.randomSplit([0.7, 0.3])
```

# Build Models

```
from pyspark.ml.classification import DecisionTreeClassifier, GBTClassifier, RandomForestClassifier
```

```
from pyspark.ml import Pipeline
```

```
dtc = DecisionTreeClassifier(labelCol='PrivateIndex', featuresCol='features')
rfc = RandomForestClassifier(numTrees=300, labelCol='PrivateIndex', featuresCol='features')
gbt = GBTClassifier(labelCol='PrivateIndex', featuresCol='features')
```

```
dtc_model = dtc.fit(train_data)
rfc_model = rfc.fit(train_data)
gbt_model = gbt.fit(train_data)
```

```
dtc_preds = dtc_model.transform(test_data)
rfc_preds = rfc_model.transform(test_data)
gbt_preds = gbt_model.transform(test_data)
```

# Evaluate Models

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator
```

```
my_binary_evaluator = BinaryClassificationEvaluator(labelCol='PrivateIndex')
```

```
print('DTC')
print(my_binary_evaluator.evaluate(dtc_preds))
```

```
print('RFC')
print(my_binary_evaluator.evaluate(rfc_preds))
```

```
print('GBT')
print(my_binary_evaluator.evaluate(gbt_preds))
```

# Multiclass Evaluator

```
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
```

```
acc_eval = MulticlassClassificationEvaluator(labelCol='PrivateIndex', metricName='accuracy')
```

```
rfc_acc = acc_eval.evaluate(rfc_preds)
```
