# K-mean Clustering Documentation Example

```
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('cluster_doc').getOrCreate()
```

```
data = spark.read.format('libsvm').load('/FileStore/tables/sample_kmeans_data.txt')
```

```
data.show()
```

```
final_data = data.select('features')
```

# Build Model

```
from pyspark.ml.clustering import KMeans
```

```
kmeans = KMeans().setK(3).setSeed(31)
```

```
model = kmeans.fit(final_data)
```

```
wssse = model.summary.trainingCost
```

```
print(wssse)
```

```
centers = model.clusterCenters()
```

```
centers
```

```
results = model.transform(final_data)
```

```
results.show()
```
