# Spark DataFrame Basic Operations

```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('ops').getOrCreate()
```

```
df = spark.read.csv('/FileStore/tables/appl_stock.csv', inferSchema=True, header=True)
```

```
df.printSchema()
```

```
df.head(5)
```

##Filter Data

### Using SQL

```
df.filter("Close < 500").select(['Open', 'Close']).show()
```

### Using Python

```
df.filter(df['Close'] < 500).select('Volume').show()
```

```
df.filter((df['Close'] < 200) &~ (df['Open']>200)).show() # The () are required
```

```
df.filter(df['Low'] == 197.16).show()
result = df.filter(df['Low'] == 197.16).collect() # Actually gives you a row object
```
